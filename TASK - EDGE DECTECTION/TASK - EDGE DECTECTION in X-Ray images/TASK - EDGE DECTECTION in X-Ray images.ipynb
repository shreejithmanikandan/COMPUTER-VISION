# READ THE IMAGE

```python
img = cv.imread('ribcage.jpeg')
plt.imshow(img)
```

```python
img = cv.cvtColor(img,cv.COLOR_RGB2BGR)
```

```python
sobelX = cv.Sobel(img,cv.CV_64F,1,0)
```

```python
sobelY = cv.Sobel(img,cv.CV_64F,0,1)
```

```python
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
```

```python
combined = cv.bitwise_or(sobelX,sobelY)
```

# LAPLACIAN OPERATOR

```python
lab = cv.Laplacian(img,cv.CV_64F,ksize = 3)
lab = np.uint8(np.absolute(lab))
```

# DISPLAY RESULT

```python
titles = ['image','Combined','Laplacian']
images = [img,combined,lab]

for i in range(3):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
cv.imshow('combined',combined)
cv.imshow('lab',lab)
cv.imwrite('combined.jpeg',combined)
cv.imwrite('laplacian.jpeg',lab)

cv.waitKey(0)
cv.destroyAllWindows()
```
