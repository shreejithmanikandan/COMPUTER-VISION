# IMPORTING LIBRARIES

```python
import numpy as np
import cv2 as cv
```

# READ IMAGE

```python
img = cv.imread("group photo.jpeg")
```

# GRAY CONVERSION

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

# READ IMAGE - TEMPLATE

```python
temp = cv.imread('crop photo.png',0)
```

# DISPLAY IMAGE

```python
cv.imshow('source',img)
cv.imshow('template',temp)

if cv.waitKey(0) & 0xff ==27:
    cv.destroyAllWinsows()
```

# TEMPLATE WIDTH AND HEIGHT

```python
print(temp.shape)
```

```python
w,h = temp.shape[::-1]
```

```python
w
```

```python
h
```

# TEMPLATE MATCHING

```python
mat = cv.matchTemplate(img_gray,temp,cv.TM_CCOEFF_NORMED)
```

# SETTING THRESHOLD

```python
threshold = 0.7
```

# MATCHED LOCATION

```python
location = np.where(mat >= threshold)
```

# MATCHING REGION

```python
for pt in zip(*location[::-1]):cv.rectangle(img,pt,(pt[0] + w, pt[1] + h),(0,255,255),2)
```

# OUTPUT DISPLAY

```python
cv.imshow("Detected",img)
if cv.waitKey(0) & 0xff ==27:
    cv.destroyAllWinsows()
```

```python
cv.imwrite("Detected.png",img)
```
